```python
# Convert 'Review Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create a new column 'Date' with the format DD:MM:YYYY
df['Formatted Date'] = df['Date'].dt.strftime('%d:%m:%Y')

# Display the first few rows to verify
print(df[['Date', 'Formatted Date']].head())
```

```python
import re
import pandas as pd
# Function to remove all emojis and special characters
def clean_text(text):
    if not isinstance(text, str):  # Check if text is a string
        return ''  # Return empty string for NaN or non-string types
    # Remove emojis
    emoji_pattern = re.compile(
        "[\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F700-\U0001F77F"  # alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251"  # Enclosed Characters
        "]+",
        flags=re.UNICODE
    )
    # Remove all emojis
    text = emoji_pattern.sub('', text)
    # Remove all special characters, keeping only alphanumeric characters and spaces
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text
```

```python
# Apply the clean_text function to the 'Review' column and save the result back to the same column
df['Reviews'] = df['Reviews'].apply(clean_text)
# Display the first few rows to verify
print(df['Reviews'].head())
```
